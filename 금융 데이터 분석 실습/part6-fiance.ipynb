```python
#part 3

import pandas as pd
df = pd.read_excel('ss_ex_1.xlsx', index_col=0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()

df["거래량"].shift(1)
```

```python
df['전일거래량'] = df['거래량'].shift(1)
df[['거래량', '전일거래량']]
```

```python
df['전일거래량'] = df['거래량'].shift(1)
cond = df['거래량'] > df['전일거래량']
df[cond]
```

```python
print('상승일:', len(df[cond]))
print('영업일:', len(df))
```

```python
df['거래량'].diff()
```

```python
cond = df['거래량'].diff()>0
len(df[cond])
```

```python
yeild = df['종가'] / df['종가'].shift(6)
cond = yeild >= 1.03
len(df[cond])
```

```python
cond.shift(1)
```

```python
cond_modified = cond.shift(1).fillna(False)
s = df.loc[cond_modified, '종가'] / df.loc[cond_modified, '시가']
s.cumprod().iloc[-1]
```

```python
# 6.4 이동평균

import pandas as pd
df = pd.read_excel('ss_ex_1.xlsx', index_col = 0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()[['종가']]

df['종가D-1']= df['종가'].shift(1)
df['종가D-2'] = df['종가'].shift(2)
df['ma3'] = (df['종가']+df['종가D-1']+df['종가D-2'])/3
df.head()
```

```python
df['rolling3'] = df['종가'].rolling(3).mean()
df.head()
```

```python
df = pd.read_excel('ss_ex_1.xlsx', index_col=0)
df.index = pd.to_datetime(df.index)

df['ma5'] = df['종가'].rolling(5).mean().shift(1)
cond = df['ma5']<df['시가']
print('상승일:', len(df[cond]))
print('영업일:', len(df))
```

```python
# 6.5 데이터 샘플링

import pandas as pd
df = pd.read_excel('ss_ex_1.xlsx', index_col=0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()[['시가', '저가', '고가', '종가', '거래량']]
df.head()
```

```python
df.resample('m').first()
```

```python
df.resample('MS').first()
```

```python
how = {
    "시가" :'first',
    "종가" : "last",
    '고가' : max,
    "저가" : min,
    "거래량" : sum,
}
df.resample('MS').agg(how)
```

```python
df.resample('3D').agg(how).head()
```

```python
temp = df.resample('3D').agg(how)
temp.index = temp.index + pd.to_timedelta('2D')
temp.dropna()
```

```python
df.resample('3D', offset='1D').agg(how)
```

```python
df.resample('12H').agg(how)
```

```python
df.resample('12H').agg(how).ffill()
```

```python
import numpy as np

temp = df.resample('12H').agg(how)
cond  = temp.isna().any(axis=1)
temp.loc[cond, '거래량'] = np.NaN
```

```python
import pandas as pd

df = pd.read_excel('ss_ex_1.xlsx', index_col=0, usecols=[0,1,4])
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df.head()
```

```python
df_quarter = df['시가'].resample('q').first().to_frame()
df_quarter
```

```python
df_quarter = df['시가'].groupby(pd.Grouper(freq='q')).first().to_frame()
df_quarter
```

```python
df_quarter['quarter'] = df_quarter.index.quarter
df['quarter'] = df.index.quarter
```

```python
df_daily = df[['종가', 'quarter']].reset_index()
r = pd.merge(left=df_daily, right=df_quarter, on='quarter')
r
```

```python
r['수익률'] = r['종가']/r['시가']
r = r.set_index(['quarter', '일자'])
```
